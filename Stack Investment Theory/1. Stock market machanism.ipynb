## 증권시장
- 한국거래소(KRX)가 운영한다. 한국거래소는 금융투자회사들의 지분의 85% 이상을 가진 주식회사(비상장회사)이다.
- 발행시장에선 상장, 발행된 주식이 유통시장에서 거래되며 기업은 자본을 조달한다.<br>
  $\bullet$ 기업은 발행시장에서 최초 상장(IPO), 추가 주식을 발행해 **기존 주주**와 거래(유상증자)하거나 무상으로 배분(무상증자)한다.<br>
  $\bullet$ 유통시장에선 주식 매도자와 주식 매수자 간 거래가 이루어진다. <br>
  $\bullet$ 발행시장에선 기업의 가치, 시장 조건, 수요 및 공급을 고려해 결정한 주식가격으로 거래되나 <br>
  $~~$ 유통시장에선 거래를 통해 각 기업의 변화, 가치 등에 발맞춰 적정주가로 거래된다.

## 매수$\cdot$매도 주문 방법
- 시장가 주문: 신속한 거래를 목적으로 현재 시장가로 **즉시** 거래를 체결하는 방법
- 지정가 주문: 특정 가격으로 거래하려는 방법
- 조건부 주문: 특정 조건(가격, 거래량)에 만족할 시 거래를 체결하려는 방법
- 옵션 주문: 주문에 추가 조건을 부여하는 주문 유형
- 장전 시장가 주문: 시장이 열린 직후 특정 주식을 빠르게 매매하려는 목적으로 시장이 열린 후 시장가로 주문이 실행되는 방법
- 장마감 시장가 주문: 시장 종료 직전에 특정 주식을 빠르게 매매하려는 목적으로 시장이 닫히기 직전에 시장가로 주문이 실행되는 방법
- 시간 유한 주문: 주문 유지 시간을 설정하는 방법

## 시장 파악을 위한 지표
- 금융시장 지수(KOSPI, KOSDAQ)<br>
  $\bullet$ 주식 시장의 흐름, 투자자들의 투자 심리, 경제 전망 등을 직접적으로 반영한 지표이다.
- 매수$\cdot$매도 스프레드( = 최저 매도가격 - 최고 매수가격)<br>
  $\bullet$ 스프레드가 넓으면 시장에서 거래가 드물게 발생하고 유동성이 낮다는 것을 나타낸다. <br>
  $\bullet$ 스프레드가 좁다면 시장에서 거래가 활발하고 유동성이 높다는 것을 나타낸다.
- 국내총생산$\cdot$ 실업률<br>
  $\bullet$ 경제의 크기와 성장률, 경기 전망을 파악하는 데 사용된다.
- 소매 판매<br>
   $\bullet$ 소매 판매량은 소비자 지출을 반영하며, 소비 동향을 파악하는 데 중요하다.
- 소비자 물가 지수<br>
  $\bullet$ 물가상승률, 그로 인한 인플레이션에 대한 신호를 파악할 수 있다.
- 금리<br>
  $\bullet$ 중앙은행의 기준 금리와 장기 및 단기 금리는 투자 환경 및 가치투자의 영향을 나타내는데 중요한 지표이다.
- 환율<br>
  $\bullet$ 국제 환율은 수출 및 수입, 외국인 투자 등에 영향을 미칠 수 있다.<br>
  $\bullet$ 수출 중심인 우리나라의 경우 중요하게 살펴볼 지표이다
